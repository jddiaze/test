# El modelo de regresión polinomial
$${\displaystyle y_{i}\,=\,\beta _{0}+\beta _{1}x_{i}+\beta _{2}x_{i}^{2}+\cdots +\beta _{m}x_{i}^{m}+\varepsilon _{i}\ (i=1,2,\dots ,n)}$$
se puede expresar en forma de matriz en términos de una matriz diseño $X$, un vector de respuesta 
$\mathbf {Y}$, un vector de parámetro ${\boldsymbol {\beta }}$ y un vector ${\boldsymbol  {\varepsilon }}$ de errores aleatorios. La i-ésima fila de $X$ y $\mathbf {Y}$ contendrá el valor $X$ e $Y$ para la muestra de datos i-ésimo. Entonces el modelo se puede escribir como un sistema de ecuaciones lineales:

$${\displaystyle {\begin{bmatrix}y_{1}\\y_{2}\\y_{3}\\\vdots \\y_{n}\end{bmatrix}}={\begin{bmatrix}1&x_{1}&x_{1}^{2}&\dots &x_{1}^{m}\\1&x_{2}&x_{2}^{2}&\dots &x_{2}^{m}\\1&x_{3}&x_{3}^{2}&\dots &x_{3}^{m}\\\vdots &\vdots &\vdots &\ddots &\vdots \\1&x_{n}&x_{n}^{2}&\dots &x_{n}^{m}\end{bmatrix}}{\begin{bmatrix}\beta _{0}\\\beta _{1}\\\beta _{2}\\\vdots \\\beta _{m}\end{bmatrix}}+{\begin{bmatrix}\varepsilon _{1}\\\varepsilon _{2}\\\varepsilon _{3}\\\vdots \\\varepsilon _{n}\end{bmatrix}},}$$
que cuando se usa la notación matricial pura se escribe como
$${\displaystyle {\boldsymbol {Y}}=X{\boldsymbol {\beta }}+{\boldsymbol {\varepsilon }}}$$
El vector de coeficientes de regresión polinomial estimados (usando la estimación de mínimos cuadrados ordinarios) es
$${\displaystyle {\widehat {\boldsymbol {\beta }}}=(X^{t}X)^{-1}X^{t}{\boldsymbol {Y}}}$$
asumiendo $m < n$ que se requiere para que la matriz sea invertible; entonces desde $X$ es una matriz de Vandermonde, se garantiza que la condición de invertibilidad se mantenga si todos los $x_{i}$ los valores son distintos. Esta es la única solución de mínimos cuadrados.

##**Suma de errores al cuadrado**
El ajuste por minimos cuadrados consiste en minimizar la suma de los cuadrados de los residuos entre la $y$ medida y la $y$ calculada según el siguiente modelo lineal (Chapra S. C. y Canale, R. P. 2007, pag 469)
$$S_r = \sum_{i=1}^{n}\varepsilon^2=\varepsilon^T\varepsilon$$

##**Estadísticos para estimar el error**
**Error estandar** (Chapra S. C. y Canale, R. P. 2007, pag 472)

$$S_{y/x} = \sqrt{\dfrac{S_r}{n - 2}}$$

donde a $s_{y/x}$ se le llama *error estandar del estimado*. El subindice "y/x" designa que el error es para un valor predicho de $y$ correspondiente a un valor particular de $x$.

**Coeficiente de Determinación** (Chapra S. C. y Canale, R. P. 2007, pag 473)

$$r^2 =1- \dfrac{S_r}{S_t}$$
 
donde $r^2$ es el *coeficiente de determinación*, $r$ es el coeficiente de correlación y $S_t$ se define en (Chapra S. C. y Canale, R. P. 2007, pag 473) como la suma total de los cuadrados de las diferencias entre los datos y la media, así

$$S_t = \sum (y_i - \bar{y})^2 $$

En un ajuste perfecto, $S_r = 0$ y $r = r2 = 1$, significa que la línea explica el 100 % de la variabilidad de los datos. Si $r = r^2 = 0$, $S_r = S_t$ el ajuste no representa alguna
mejora.

1. Use regresión lineal para determinar un modelo de predecir $c$ como función de $T$,

2. Use regresión cuadrática para determinar un modelo de predecir $c$ como función de $T$,

3. Encuentre la suma de los errores al cuadrado para las dos situaciones (lineal y cuadrática).

4. Haga representación gráfica y haga una predicción de $c$ para $T=2000$ en las dos situaciones (lineal y cuadrática).

5. Encuentre el error estandar $S_{y/x}$.

6. Encuentre el coeficiente de determinación $r^2$, para los dos situaciones (lineal y cuadrático).

7. ¿Qué modelo se ajusta mejor a los datos?


Solución paso 1

```python
from sympy import *
init_printing()
Y,X,B,n,y,a,b,b1,b2,x1,x2=symbols('Y X B n y a b b_1 b_2 x_1 x_2')
import matplotlib.pyplot as plt
```

```python
# Construir el vector x
x = Matrix([-50,-30,0,60,90,110])
x
```

```python
# Construir el vector Y
Y = Matrix([1270,1280,1350,1480,1580,1700])
Y
```

```python
# Realizar la gráfica de los datos
plt.figure()
plt.plot(x,Y,"o")
plt.show()
```

```python
# Construir el vector de unos
u = ones(6,1)
u
```

```python
# Construir la matriz X
X = Matrix([[u,x]])
X
```

```python
# Constrir el vector B
B = Matrix([1,b])
B
```

```python
# Construir los parámetros de la solución
S = (X.T*X).inv()*X.T*Y
S
```

```python
# Construir la ecuacíon de la recta de regresión
Eq(y,(S.T*B)[0])
```

```python
# Construir la gráfica de la recta de regresión
import numpy as np
plt.figure()
plt.plot(x,Y,"o")
plt.plot(x,S[0]+S[1]*np.array(x))
plt.show()
```

Solución paso 2

```python
# Construir el vector x
XC = Matrix([(-50)**2,(-30)**2,0**2,60**2,90**2,110**2])
XC
```

```python
# Realizar la gráfica de los datos
plt.figure()
plt.plot(x,Y,"o")
plt.show()
```

```python
# Construir la matriz X
XC = Matrix([[u,x,XC]])
XC
```

```python
# Constrir el vector B
BC = Matrix([1,b,b**2])
BC
```

```python
# Construir los parámetros de la solución
Sc = (XC.T*XC).inv()*XC.T*Y
Sc
```

```python
# Construir la ecuacíon de la recta de regresión
Eq(y,(Sc.T*BC)[0])
```

```python
# Construir la gráfica de la recta de regresión
plt.figure()
plt.plot(x,Y,"o")
plt.plot(x,Sc[0]+Sc[1]*np.array(x)+Sc[2]*np.array(x)**2)
plt.show()
```

Solución paso 3

```python
el=Y-(X*S)
el1 = (el.T)*el
print("La suma de los errores al cuadrado del modelo lineal es: ",el1[0].evalf())
```

```python
ec=Y-(XC*Sc)
ec1 = (ec.T)*ec
print("La suma de los errores al cuadrado del modelo cuadrático es: ",ec1[0].evalf())
```

```python
ec1[0] < el1[0]
```

Solución paso 4

```python
ecuacion_lineal = Eq(y,(S.T*B)[0])
ecuacion_lineal_evaluada = ecuacion_lineal.subs({b: 2000})
subs_l = ecuacion_lineal_evaluada.evalf()
subs_l
```

```python
ecuacion_cuadratica = Eq(y,(S.T*B)[0])
ecuacion_cuadratica_evaluada = ecuacion_cuadratica.subs({b: 2000})
subs_c = ecuacion_cuadratica_evaluada.evalf()
subs_c
```

Solución paso 5

```python
n = len(x)
SER_l = sqrt(el1 / (n - 2))
SER_c = sqrt(ec1 / (n - 2))
print("El error estándar del modelo lineal es: ",SER_l.evalf())
print("El error estándar del modelo cuadrático es: ",SER_c.evalf())
SER_l
```

Solución paso 6

```python
# Calcular St
import sympy
Y_mean = sympy.Add(*Y)/n
St = sum([(Y[i] - Y_mean)**2 for i in range(n)])

# Calcular r^2
r_squared_l = 1 - (el1[0] / St)
r_squared_c = 1 - (ec1[0] / St)
print('El coeficiente de determinación del modelo lineal es:', r_squared_l.evalf())
print('El coeficiente de determinación del modelo cuadrático es:', r_squared_c.evalf())
```

Solución paso 7

El mejor modelo es el modelo cuadrático, ya que su coeficiente de determinación es más cercado a 1.
